```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
df = pd.read_excel('Updated_From_DEV_Clean_1.xlsx')
```

```python

fig = go.Figure()


for _, row in df.iterrows():
    if row['LS-Latitude'] != 0.0 and row['LS-Longitude'] != 0.0 and row['F-Latitude'] != 0.0 and row['F-Longitude'] != 0.0:
        fig.add_trace(go.Scattermapbox(
            mode='lines+markers',
            lat=[row['LS-Latitude'], row['F-Latitude']],
            lon=[row['LS-Longitude'], row['F-Longitude']],
            line=dict(width=2, color='blue'),
            marker=dict(size=8),
            text=[row['Name'], 'Destination'], 
            name=row['Name'] 
        ))


fig.add_trace(go.Scattermapbox(
    mode='markers',
    lat=df['LS-Latitude'].tolist() + df['F-Latitude'].tolist(),
    lon=df['LS-Longitude'].tolist() + df['F-Longitude'].tolist(),
    marker=dict(size=8, color='red'),
    text=df['Name'].tolist() + ['Destination']*len(df), 
    name='Locations'
))


fig.update_layout(
    mapbox=dict(
        center={"lat": 56.1304, "lon": -106.3468},
        zoom=2,
        style="carto-positron"
    ),
    title="Map with Lines Between Locations"
)


fig.show()
```

```python

filtered_df = df[
    (df['LS-Latitude'] != 0.0) & 
    (df['LS-Longitude'] != 0.0) & 
    (df['F-Latitude'] != 0.0) & 
    (df['F-Longitude'] != 0.0)
]


fig = go.Figure()


for _, row in filtered_df.iterrows():
    fig.add_trace(go.Scattermapbox(
        mode='lines',
        lat=[row['LS-Latitude'], row['F-Latitude']],
        lon=[row['LS-Longitude'], row['F-Longitude']],
        line=dict(width=2, color='lightblue'), 
        name=row['Name']  
    ))


fig.add_trace(go.Scattermapbox(
    mode='markers',
    lat=[lat if lat != 0.0 else None for lat in df['LS-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['LS-Latitude'], df['LS-Longitude'])],
    marker=dict(size=8, color=['grey' if value == 'Missing' else 'blue' for value in df['Status']]),
    text=[f"Name: {name}<br>Status: {status}<br>Last Seen Location: {loc}<br>Last Seen Location Source: {locSrc}" 
          for name, status, loc,locSrc in zip(df['Name'], df['Status'], df['LS-General-Location'], df['LS-Location Source'])],  
    hoverinfo='text',   
    name='Last Seen Locations'
))

fig.add_trace(go.Scattermapbox(
    mode='markers',
    lat=[lat if lat != 0.0 else None for lat in df['F-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['F-Latitude'], df['F-Longitude'])],
    marker=dict(size=8, color=['red' if value == 'Murdered' else 'orange' for value in df['Status']]),
    text=[f"Name: {name}<br>Status: {status}<br>Found Location: {loc}<br>Found Location Source: {locSrc}" 
          for name, status, loc, locSrc in zip(df['Name'], df['Status'], df['F-General-Location'],df['F-Location Source'])],  
    hoverinfo='text',  
    name='Found Locations'
))

fig.update_layout(
    mapbox=dict(
        center={"lat": 56.1304, "lon": -106.3468},
        zoom=2,
        style="carto-positron"
    ),
    title="Map with Lines Between Last Seen and Found Locations",
    height=800,  
    width=1200   
)

# Show the plot
fig.show()
```

```python
import plotly.express as px
import pandas as pd


filtered_df = df[
    (df['LS-Latitude'] != 0.0) & 
    (df['LS-Longitude'] != 0.0) & 
    (df['F-Latitude'] != 0.0) & 
    (df['F-Longitude'] != 0.0)
]


lines_df = filtered_df.copy()
lines_df['lat'] = list(zip(filtered_df['LS-Latitude'], filtered_df['F-Latitude']))
lines_df['lon'] = list(zip(filtered_df['LS-Longitude'], filtered_df['F-Longitude']))


fig = px.scatter_map(df,
    lat=[lat if lat != 0.0 else None for lat in df['LS-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['LS-Latitude'], df['LS-Longitude'])],
    color=['grey' if value == 'Missing' else 'blue' for value in df['Status']],
    size_max=8,
    hover_name=df['Name'],
    hover_data={'Name': False, 'Status': True, 'LS-General-Location': True, 'LS-Location Source': True},
    title="Map with Lines Between Last Seen and Found Locations"
)


for _, row in lines_df.iterrows():
    fig.add_scattermapbox(
        mode='lines',
        lat=row['lat'],
        lon=row['lon'],
        line=dict(width=2, color='purple'),
        name=row['Name']
    )


fig.add_scattermapbox(
    lat=[lat if lat != 0.0 else None for lat in df['F-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['F-Latitude'], df['F-Longitude'])],
    marker=dict(size=8, color=['red' if value == 'Murdered' else 'orange' for value in df['Status']]),
    text=[f"Name: {name}<br>Status: {status}<br>Found Location: {loc}<br>Found Location Source: {locSrc}" 
          for name, status, loc, locSrc in zip(df['Name'], df['Status'], df['F-General-Location'], df['F-Location Source'])],  
    hoverinfo='text',
    name='Found Locations'
)


fig.update_layout(
    mapbox=dict(
        center={"lat": 56.1304, "lon": -106.3468},
        zoom=2,
        style="open-street-map"
    ),
    height=800,
    width=1200
)


fig.show()
```
