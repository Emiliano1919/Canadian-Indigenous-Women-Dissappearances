```python
import plotly.express as px
import pandas as pd

df = pd.read_excel('Updated_From_DEV_Clean_1.xlsx')

df.head()
```

```python
fig = px.scatter_mapbox(
    df,
    lat='LS-Latitude',
    lon='LS-Longitude',
    hover_name='Name',  
    title='Map of Last Seen Locations',
    mapbox_style='carto-positron',  
    zoom=2, 
    center={"lat": 56.1304, "lon": -106.3468},
)

# Show the plot
fig.show()
```

```python

fig = px.scatter_mapbox(
    df,
    lat='F-Latitude',
    lon='F-Longitude',
    hover_name='Name',  
    title='Map of Found Locations',
    mapbox_style='carto-positron', 
    zoom=2, 
    center={"lat": 56.1304, "lon": -106.3468},
)

fig.show()
```

```python

filtered_df = df[
    (df['LS-Latitude'] != 0.0) & 
    (df['LS-Longitude'] != 0.0) & 
    (df['F-Latitude'] != 0.0) & 
    (df['F-Longitude'] != 0.0)
]


fig = go.Figure()


for _, row in filtered_df.iterrows():
    fig.add_trace(go.Scattermapbox(
        mode='lines',
        lat=[row['LS-Latitude'], row['F-Latitude']],
        lon=[row['LS-Longitude'], row['F-Longitude']],
        line=dict(width=2, color='blue'),  # Line color
        name=row['Name']  # Line legend
    ))


fig.add_trace(go.Scattermapbox(
    mode='markers',
    lat=[lat if lat != 0.0 else None for lat in df['LS-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['LS-Latitude'], df['LS-Longitude'])],
    marker=dict(size=8, color=['grey' if value == 'Missing' else 'blue' for value in df['Status']]),
    text=[name if lat != 0.0 else 'Invalid' for lat, name in zip(df['LS-Latitude'], df['Name'])],  
    name='Last Seen Locations'
))

fig.add_trace(go.Scattermapbox(
    mode='markers',
    lat=[lat if lat != 0.0 else None for lat in df['F-Latitude']],
    lon=[lon if lat != 0.0 else None for lat, lon in zip(df['F-Latitude'], df['F-Longitude'])],
    marker=dict(size=8, color=['red' if value == 'Murdered' else 'orange' for value in df['Status']]),
    text=['Invalid' if lat == 0.0 else 'Found dead' for lat in df['F-Latitude']],  # Optional hover text
    name='Found Locations'
))

fig.update_layout(
    mapbox=dict(
        center={"lat": 56.1304, "lon": -106.3468},
        zoom=2,
        style="carto-positron"
    ),
    title="Map with Lines Between Last Seen and Found Locations",
    height=800,  # Adjust the height of the map
    width=1200   # Adjust the width of the map
)

# Show the plot
fig.show()
```
