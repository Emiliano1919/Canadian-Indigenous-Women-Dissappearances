```python
import pandas as pd
```

```python
df = pd.read_excel('Updated_From_DEV.xlsx') 
print(df.columns)
df
```

```python
df.iloc[18]
```

```python
start_index = 19
mask = df.index >= start_index
df.loc[mask, 'LS-Latitude'] = df.loc[mask, 'LS-Location'].str.split(',', expand=True)[0].str.strip().astype(float)
df.loc[mask, 'LS-Longitude'] = df.loc[mask, 'LS-Location'].str.split(',', expand=True)[1].str.strip().astype(float)
print(df.head(24))
```

```python
start_index = 19
mask = df.index >= start_index
df.loc[mask, 'F-Latitude'] = df.loc[mask, 'F-Location'].str.split(',', expand=True)[0].str.strip().astype(float)
df.loc[mask, 'F-Longitude'] = df.loc[mask, 'F-Location'].str.split(',', expand=True)[1].str.strip().astype(float)
print(df.head(24))
```

```python
finishing_index = 18
mask = df.index <= finishing_index
df.loc[mask, 'LS-Location'] = df.loc[mask,'LS-Latitude'].astype(str) + ',' + df.loc[mask,'LS-Longitude'].astype(str)
df.loc[15:25]
```

```python
finishing_index = 18
mask = df.index <= finishing_index
df.loc[mask, 'F-Location'] = df.loc[mask,'F-Latitude'].astype(str) + ',' + df.loc[mask,'F-Longitude'].astype(str)
df.loc[15:25]
```

```python
df['LS-Latitude'] = df['LS-Latitude'].fillna(0.0)
df['LS-Longitude'] = df['LS-Longitude'].fillna(0.0)
df['F-Latitude'] = df['F-Latitude'].fillna(0.0)
df['F-Longitude'] = df['F-Longitude'].fillna(0.0)
df['F-Location'] = df['F-Location'].fillna(0.0)
df['LS-Longitude'] = df['LS-Longitude'].fillna(0.0)
```
